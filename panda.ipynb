```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
from multiprocessing import Pool
import csv

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
browser.get('https://www.panda.tv/10003')
```

```python
chunk = browser.find_elements_by_xpath('//li[@class="room-chat-item room-chat-message     "]')
chunk[-1].find_element_by_xpath('.//span[@class="room-chat-content"]').text
```

```python
while True:
    chunk = browser.find_elements_by_xpath('//li[@class="room-chat-item room-chat-message     "]')
    wait = WebDriverWait(browser, 3600)
    
    try:
        node = chunk[-1]        
        for danmu in chunk:
            cont = danmu.find_element_by_xpath('.//span[@class="room-chat-content"]').text
            print(cont,end=',')
        wait.until(EC.staleness_of(node))
    except:
        print('Stale DOM occured!')
        time.sleep(.5)
        pass
```
